Topics: .loc and .iloc

 - .loc selects rows and columns by label/name

 - .iloc selects row and columns by index

```python
# .loc structure: .loc[rows (*must provide a range here*), [*column names here*]]
movies_df.loc[2:4, ['Title','Director','Year'] ]
```

```python
# .iloc structure: .iloc[rows (*must provide a range based on index here*), [*column number(s) corresponding the column name here*]]
movies_df.iloc[ :, [1,2,3] ]
  # placing a colon will select all rows or columns, depending on where you place it
```

```python
# .value_counts() counts each of the unique values contained in the column(s) you identify

movies_df.loc[:,'Director'].value_counts()
# this line of code locates the column name(s) you want to focus on, then get the count of each unique values in that column
```

```python
# .mean() gets the average of the column(s) you select. this only works for columns that have numerical values
movies_df.loc[:, 'Length_minutes'].mean()
```

```python
# .groupby aggregates your numbers by the column(s) you select

movies_df.loc[:, ['Director', 'Length_minutes']].groupby('Director').mean()
  # this gets the average length of minutes for each director
  # REMEMBER: you must also state the column you want to group by within the .loc function as well
```

Topics: Filtering and Sorting Data

```python
# you can filter data in 2 ways; please run both methods separately!!

# Method 1: using .loc and operators + creating a filter within a filter
  # 1a:
  # locate all the rows whose length in minutes are greater than 100
movie_filter = movies_df.loc[:, "Length_minutes"] > 100

  # then use .loc again where you select the filter you created initially, and take all the column names
movies_df.loc[movie_filter, :]

  # 1b:
  # orrrr both can be combined into this as well:
movies_df.loc[movies_df.loc[:, "Length_minutes"] > 100, :]

# Method 2: using .query
movie_filter2 = movies_df.query("Length_minutes > 100")
movie_filter2
```

```python
# you can have multiple conditions in your filtering

# Method 1: using .loc and operators + creating a filter within a filter
movie_len_filter = movies_df.loc[:, "Length_minutes"] > 100
movie_year_filter = movies_df.loc[:, "Year"] < 2005

movies_df.loc[(movie_len_filter) & (movie_year_filter), :]

# Method 2: using .query
movie_filter3 = movies_df.query("Length_minutes > 100 & Year < 2005")
movie_filter3
```

```python
# sort_values() sorts your data
# your data gets sorted ascending by default; you must place ascending=False if you want to sort your data in descending order

movies_df.loc[:,"Title"].sort_values().reset_index(drop=True)
  # .reset_index(drop=True) re-assigns the indexes after sorting
```

```python
movies_df.sort_values("Title").reset_index(drop=True)
# this sorts your data by title in an ascending order, then resets the index
```

```python
# you can sort by multiple columns

# follow this structure:
  # dataframe_name.sort_values([Col1 name, Col2 name], ascending=[True or False for Col1, True or False for Col2]).reset_index(drop=True)

movies_df.sort_values(["Director","Title"], ascending=[True, False]).reset_index(drop=True)
```

Topics: Concat and merge

- Concat puts 2 tables together whether both have a common column to join on or not

- Merge puts 2 data together under a common column. Think of how joins work in SQL

```python
# we first import a new data :))

other_movies_url = "https://raw.githubusercontent.com/freestackinitiative/COOP-PythonLessons/main/lessons/lesson3/data/Other_Movies.csv"
other_movies_df = pd.read_csv(other_movies_url)
```

```python
# then load it here:
other_movies_df.head()
```

```python
# pd.concat() puts 2 tables together; the second table/data you specify gets added into the next row by default
  # observe when you run the next block of code

all_movies_df = pd.concat([movies_df,other_movies_df]).reset_index(drop=True)
```

```python
# creating new data using dictionaries to later portray how .merge() function is used
managers = pd.DataFrame(
    {
    'Id': [1,2,3],
    'Manager':['Chris','Maritza','Jamin']
    }
)

managers.head()
```

```python
captains = pd.DataFrame(
    {
    'Id': [2,2,3,1,1,3,2,3,1,1,3,3],
    'Captain':['Derick','Shane','Becca','Anna','Christine','Melody','Tom','Eric','Naomi','Angelina','Nancy','Richard'],
    'Title':['C','C','SC','C','SC','C','C','SC','C','EC','C','SC']
    }
)

captains.head(12)
```

```python
# .merge() joins 2 tables on a column both have in common; again, think of joins from SQL

# follow this structure: new_table = table1.merge(table2, left_on = 'column from table1', right_on = 'column from table2')

roster = captains.merge(managers,left_on = 'Id', right_on = 'Id')
roster
```

```python
# bonus: you can add a 'how = ___' after 'right_on' to specify how you want the join to occur
  # like in SQL, you can do inner, left, right, and outer joins in python; then place them in your how statement

# follow this structure:
  # new_table = table1.merge(table2, left_on = 'column from table1', right_on = 'column from table2', how = 'inner, left, right, or outer')

# this is how it's applied in our current scenario:
roster1 = captains.merge(managers,left_on = 'Id', right_on = 'Id', how='inner')
roster1
```

```python
# bringing back concat once again to portray differences of concat and merge

test_roster = pd.concat([captains, managers], axis="columns").reset_index(drop=True)
test_roster.head()
```

Topics: Handling nulls

```python
cars_url = "https://raw.githubusercontent.com/freestackinitiative/COOP-PythonLessons/main/lessons/lesson3/data/Cars.csv"
cars = pd.read_csv(cars_url)
cars.head(3)
```

```python
# sort data in alphabetical order
cars = cars.sort_values("Company").reset_index(drop=True)
cars
```

```python
# .isna().sum() counts the amount of nulls of each column in your data

cars.isna().sum()
```

```python
# since location and year have nulls, we will dig further into the data

# the below code filters the data where Location is null

missing_car_info_filter = cars.loc[:, "Location"].isna()
cars.loc[missing_car_info_filter, :]
```

```python
# the below code filters the data where Year is null

cars.loc[cars.loc[:, "Year"].isna(), :]
```

```python
# nulls can be dealt with in various ways:

# method 1: filling it in with a particular value
  # in this case, we're filling in nulls in the Location column with NA

cars.loc[:, "Location"] = cars.loc[:, "Location"].fillna(value="NA")
cars
```

```python
# recheck if nulls still exist. having this in place from time to time is best practice when cleaning and manipulating data with nulls
cars.isna().sum()
```

```python
# method 2: dropping the rows that have null values
  # below code drops the rows whose Year values are null

cars = cars.dropna().reset_index(drop=True)
cars
```

```python
# recheck nulls one last time

cars.isna().sum()
```

```python
# another way of checking nulls is using .info()
cars.info()
```
